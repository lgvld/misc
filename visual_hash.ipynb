```python
import codecs
import hashlib

import matplotlib.pyplot as plt
import numpy as np

def sha256(string):
    return hashlib.sha256(string.encode()).hexdigest()

def values(string):
    hash = hashlib.sha256(string.encode()).hexdigest()
    values = list(hash)
    values = list(map(ord, values))
    # return np.asarray(values) - ord("0")
    return [v - ord("0") for v in values]

STRING = "test"
print(sha256(STRING))
print(values(STRING))
```

```python
STRING = "louis"

v = values(STRING)
v = np.reshape(v, (8,8))
print(v)
plt.imshow(v)
plt.colorbar()
plt.show()
```
